6. What is a discrete uniform distribution

  --> The discrete uniform distribution is a type of probability distribution where each possible outcome has an equal probability of occurring. It is used when there are a finite number of equally likely outcomes.

7. What are the key properties of a Bernoulli distribution

  --> The Bernoulli distribution is the simplest probability distribution, representing a single experiment with two possible outcomes: success (1) and failure (0). It is widely used in probability and statistics.

8. What is the binomial distribution, and how is it used in probability

  --> The Binomial distribution models the number of successes in a fixed number of independent Bernoulli trials, where each trial has only two possible outcomes: success (1) or failure (0).

9. What is the Poisson distribution and where is it applied

  --> The Poisson distribution models the probability of a certain number of events occurring in a fixed interval of time or space, assuming the events happen independently and at a constant average rate. It is commonly used for rare events.

10. What is a continuous uniform distribution

  --> The Continuous Uniform Distribution is a probability distribution where all outcomes in a given range are equally likely. It is often used when there is no preference or bias toward any particular value within the range.

11. What are the characteristics of a normal distribution

  --> The Normal Distribution (also called the Gaussian distribution) is a continuous probability distribution that is symmetric around its mean. It is the most important distribution in probability and statistics because of its role in the Central Limit Theorem.



12. What is the standard normal distribution, and why is it important

  -->  The Standard Normal Distribution is a special case of the Normal Distribution where the mean is 0 and the standard deviation is 1. It is the foundation for statistical analysis and hypothesis testing.



13. What is the Central Limit Theorem (CLT), and why is it critical in statistics

  --> The Central Limit Theorem (CLT) states that the distribution of the sample mean (or sum) of a sufficiently large number of independent, identically distributed (i.i.d.) random variables approaches a normal distribution, regardless of the original distribution of the data.

14. How does the Central Limit Theorem relate to the normal distribution

  --> The Central Limit Theorem (CLT) is a fundamental concept in statistics that explains why the normal distribution is so commonly observed in real-world data. It states that when taking sufficiently large samples from any population, the distribution of the sample means (or sums) will tend to be normal, regardless of the shape of the original population distribution.

15. What is the application of Z statistics in hypothesis testing

  --> Z-statistics (Z-tests) are used in hypothesis testing when the population variance is known and the data follows a normal distribution (or the sample size is large enough for the Central Limit Theorem to apply).



16. How do you calculate a Z-score, and what does it represent

  --> A Z-score (Standard Score) tells us how many standard deviations a data point is from the mean of a dataset. It helps compare values from different distributions and determine outliers.

17. What are point estimates and interval estimates in statistics

  --> A point estimate is a single value used to estimate an unknown population parameter. It is calculated from sample data.

  An interval estimate provides a range of values within which the population parameter is likely to lie. It includes a margin of error to account for uncertainty.

18. What is the significance of confidence intervals in statistical analysis

  --> A Confidence Interval (CI) is a range of values used to estimate an unknown population parameter with a certain level of confidence. Instead of relying on a single number (point estimate), confidence intervals provide a range that accounts for sampling variability and uncertainty.



19. What is the relationship between a Z-score and a confidence interval

  --> The Z-score and the Confidence Interval (CI) are closely related in statistical analysis, as the Z-score determines the margin of error in confidence interval calculations.



20. How are Z-scores used to compare different distributions

  --> A Z-score (standard score) is a statistical measure that expresses how far a data point is from the mean of its distribution in terms of standard deviations. It allows us to compare values from different distributions, even if they have different means and standard deviations.



21. What are the assumptions for applying the Central Limit Theorem

  --> The Central Limit Theorem (CLT) states that the sampling distribution of the sample mean (or sum) approaches a normal distribution, regardless of the original population’s shape, as long as the sample size is sufficiently large.

22. What is the concept of expected value in a probability distribution

  --> The expected value (EV) of a probability distribution is the long-run average outcome of a random variable over many trials. It represents the theoretical mean of the distribution and provides a measure of the "center" of the possible outcomes.

23. How does a probability distribution relate to the expected outcome of a random variable

  --> A probability distribution describes the possible values of a random variable and their associated probabilities. The expected value (EV) of the random variable is the theoretical average outcome when the experiment is repeated many times.

```python
# 2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import randint

# Define the discrete uniform distribution parameters
low, high = 1, 6  # Equivalent to a fair die (values between 1 and 6)

# Generate 10,000 random samples from the distribution
samples = randint.rvs(low, high + 1, size=10000)

# Calculate the PMF
x_values = np.arange(low, high + 1)
pmf_values = [1 / (high - low + 1)] * len(x_values)  # All values have equal probability

# Plot the PMF
plt.figure(figsize=(8, 5))
sns.barplot(x=x_values, y=pmf_values, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel("Value of X")
plt.ylabel("Probability (PMF)")
plt.title("PMF of a Discrete Uniform Distribution (Die Roll)")
plt.ylim(0, max(pmf_values) + 0.05)

# Show the plot
plt.show()
```

```python
# 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

def bernoulli_pmf(x, p):
    """
    Compute the Probability Mass Function (PMF) of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (0 or 1).
    p (float): Probability of success (0 ≤ p ≤ 1).

    Returns:
    float: Probability of x occurring.
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 in a Bernoulli distribution")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1 (inclusive)")

    return (p ** x) * ((1 - p) ** (1 - x))

# Example usage
p_success = 0.6  # Probability of success
print(f"P(X=0): {bernoulli_pmf(0, p_success):.4f}")
print(f"P(X=1): {bernoulli_pmf(1, p_success):.4f}")
```

```python
# 4. Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom

# Parameters for the Binomial distribution
n = 10    # Number of trials
p = 0.5   # Probability of success
size = 10000  # Number of simulations

# Generate random binomially distributed numbers
samples = np.random.binomial(n, p, size)

# Plot histogram
plt.figure(figsize=(8, 5))
sns.histplot(samples, bins=np.arange(n + 2) - 0.5, kde=False, color='skyblue', edgecolor='black')

# Overlay theoretical PMF
x_values = np.arange(0, n + 1)
pmf_values = binom.pmf(x_values, n, p)
plt.plot(x_values, pmf_values * size, 'ro-', label="Theoretical PMF")  # Scale PMF to match histogram

# Labels and title
plt.xlabel("Number of Successes (k)")
plt.ylabel("Frequency")
plt.title(f"Binomial Distribution Histogram (n={n}, p={p})")
plt.xticks(range(n + 1))
plt.legend()

# Show plot
plt.show()
```

```python
# 9. Write a Python function to calculate Z-scores from a dataset and plot them

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_z_scores(data):
    """
    Calculate Z-scores for a dataset.

    Parameters:
    data (array-like): Input dataset.

    Returns:
    numpy array: Z-scores for each data point.
    """
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # Using sample standard deviation
    z_scores = (data - mean) / std_dev
    return z_scores

def plot_z_scores(data):
    """
    Calculate and plot the Z-scores of a dataset.

    Parameters:
    data (array-like): Input dataset.
    """
    z_scores = calculate_z_scores(data)

    # Plot histogram of Z-scores
    plt.figure(figsize=(8, 5))
    sns.histplot(z_scores, bins=30, kde=True, color='skyblue', edgecolor='black', label="Z-scores")

    # Plot standard normal curve for reference
    x_values = np.linspace(-4, 4, 100)
    y_values = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x_values**2)
    plt.plot(x_values, y_values, 'r-', lw=2, label="Standard Normal PDF")

    # Labels and title
    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.title("Z-score Distribution")
    plt.axvline(0, color='black', linestyle='--', label="Mean (Z=0)")
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.7)

    # Show plot
    plt.show()

# Example Usage
data = np.random.normal(50, 10, 1000)  # Generate 1000 data points from N(50,10)
plot_z_scores(data)
```

```python
# 10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate a non-normal population (Exponential distribution)
population_size = 100000
population = np.random.exponential(scale=2, size=population_size)  # Exponential(λ=1/2)

# Function to sample means and demonstrate CLT
def demonstrate_clt(sample_size, num_samples=1000):
    """
    Demonstrate the Central Limit Theorem by sampling means from a non-normal population.

    Parameters:
    sample_size (int): Number of observations in each sample.
    num_samples (int): Number of samples to draw.

    Returns:
    None
    """
    sample_means = []  # Store sample means

    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=False)
        sample_means.append(np.mean(sample))

    # Plot histogram of sample means
    plt.figure(figsize=(8, 5))
    sns.histplot(sample_means, bins=30, kde=True, color='skyblue', edgecolor='black', label=f"Sample Size = {sample_size}")

    # Labels and title
    plt.xlabel("Sample Mean")
    plt.ylabel("Frequency")
    plt.title(f"Sampling Distribution of the Mean (n={sample_size})")
    plt.axvline(np.mean(population), color='red', linestyle='--', label="Population Mean")
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.7)

    # Show plot
    plt.show()

# Step 2: Apply CLT for different sample sizes
for sample_size in [5, 30, 100]:
    demonstrate_clt(sample_size)
```

```python
# 15. Simulate multiple samples from a normal distribution and verify the Central Limit Theorem

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Step 1: Generate a normal population
population_mean = 50
population_std = 15
population_size = 100000  # Large population
population = np.random.normal(population_mean, population_std, population_size)

# Function to demonstrate CLT with normal distribution
def verify_clt(sample_size, num_samples=1000):
    """
    Draw multiple samples from a normal population and verify the CLT.

    Parameters:
    sample_size (int): Number of observations in each sample.
    num_samples (int): Number of samples to draw.

    Returns:
    None
    """
    sample_means = []  # Store sample means

    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=False)
        sample_means.append(np.mean(sample))

    # Step 2: Plot histogram of sample means
    plt.figure(figsize=(8, 5))
    sns.histplot(sample_means, bins=30, kde=True, color='skyblue', edgecolor='black', label=f"Sample Size = {sample_size}")

    # Overlay theoretical normal distribution
    theoretical_mean = population_mean
    theoretical_std = population_std / np.sqrt(sample_size)  # Standard error
    x_values = np.linspace(min(sample_means), max(sample_means), 100)
    pdf_values = norm.pdf(x_values, theoretical_mean, theoretical_std)
    plt.plot(x_values, pdf_values, 'r-', lw=2, label="Theoretical Normal Curve")

    # Labels and title
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
    plt.title(f"Sampling Distribution of the Mean (n={sample_size})")
    plt.axvline(population_mean, color='red', linestyle='--', label="Population Mean")
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.7)

    # Show plot
    plt.show()

# Step 3: Apply CLT for different sample sizes
for sample_size in [5, 30, 100]:
    verify_clt(sample_size)
```

```python
# 18. import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

def calculate_z_score(data_point, data):
    """
    Calculate the Z-score of a given data point.

    Parameters:
    data_point (float): The value to compute the Z-score for.
    data (array-like): The dataset.

    Returns:
    float: The computed Z-score.
    """
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # Using sample standard deviation
    z_score = (data_point - mean) / std_dev
    return z_score

def compare_z_score(data_point, data):
    """
    Calculate the Z-score of a given data point and visualize it compared to a standard normal distribution.

    Parameters:
    data_point (float): The value to compute the Z-score for.
    data (array-like): The dataset.

    Returns:
    None
    """
    z_score = calculate_z_score(data_point, data)

    # Step 1: Plot Standard Normal Distribution
    x_values = np.linspace(-4, 4, 1000)
    pdf_values = norm.pdf(x_values, 0, 1)

    plt.figure(figsize=(8, 5))
    plt.plot(x_values, pdf_values, 'r-', lw=2, label="Standard Normal PDF (μ=0, σ=1)")

    # Step 2: Highlight the Z-score on the distribution
    plt.axvline(z_score, color='blue', linestyle='--', linewidth=2, label=f"Z-score = {z_score:.2f}")

    # Step 3: Fill the probability area under the curve
    x_fill = np.linspace(-4, z_score, 500)
    plt.fill_between(x_fill, norm.pdf(x_fill, 0, 1), alpha=0.3, color='blue', label="Probability Area")

    # Labels and title
    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.title(f"Comparison of Z-score ({z_score:.2f}) with Standard Normal Distribution")
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.7)

    # Show plot
    plt.show()

# Example Usage
data = np.random.normal(50, 10, 1000)  # Generate 1000 data points from N(50,10)
data_point = 65  # Example value to compare
compare_z_score(data_point, data)
```

```python
# 19. Implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
import scipy.stats as stats

def z_test(sample, pop_mean, pop_std, alpha=0.05, tail='two'):
    """
    Performs a Z-test for a sample dataset.

    Parameters:
    sample (array-like): Sample data
    pop_mean (float): Population mean (H0)
    pop_std (float): Population standard deviation (σ)
    alpha (float): Significance level (default 0.05)
    tail (str): Type of test - 'two' (default), 'left', or 'right'

    Returns:
    None (Prints the result of the hypothesis test)
    """
    sample_size = len(sample)
    sample_mean = np.mean(sample)

    # Calculate the Z-score
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

    # Compute the p-value based on test type
    if tail == 'two':
        p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test
    elif tail == 'left':
        p_value = stats.norm.cdf(z_score)  # Left-tailed test
    elif tail == 'right':
        p_value = 1 - stats.norm.cdf(z_score)  # Right-tailed test
    else:
        raise ValueError("Invalid tail argument. Choose 'two', 'left', or 'right'.")

    # Print results
    print(f"Sample Mean: {sample_mean:.3f}")
    print(f"Z-score: {z_score:.3f}")
    print(f"P-value: {p_value:.5f}")

    # Decision Rule
    if p_value < alpha:
        print("Result: Reject the null hypothesis (H0)")
    else:
        print("Result: Fail to reject the null hypothesis (H0)")

# Example Usage
np.random.seed(42)
sample_data = np.random.normal(52, 10, 50)  # Sample from N(52,10), n=50
population_mean = 50  # H0: μ = 50
population_std = 10   # Known σ

# Perform a two-tailed Z-test
z_test(sample_data, pop_mean=population_mean, pop_std=population_std, tail='two')
```

```python
# 20. Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    """
    Calculates and prints the confidence interval for a dataset.

    Parameters:
    data (array-like): Sample dataset
    confidence (float): Confidence level (default 0.95)

    Returns:
    tuple: (Lower bound, Upper bound)
    """
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error = s / sqrt(n)

    # Determine t-score for small samples, or Z-score for large samples
    if n >= 30:
        critical_value = stats.norm.ppf(1 - (1 - confidence) / 2)  # Z-score
    else:
        critical_value = stats.t.ppf(1 - (1 - confidence) / 2, df=n-1)  # t-score

    margin_of_error = critical_value * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    print(f"Sample Mean: {mean:.3f}")
    print(f"Confidence Level: {confidence*100}%")
    print(f"Margin of Error: {margin_of_error:.3f}")
    print(f"Confidence Interval: ({lower_bound:.3f}, {upper_bound:.3f})")

    return lower_bound, upper_bound

# Example Usage
np.random.seed(42)
sample_data = np.random.normal(100, 15, 50)  # Sample from N(100,15), n=50
confidence_interval(sample_data, confidence=0.95)
```

```python
# 21. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Generate Normal Distribution Data
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=50)  # N(100, 15), n=50

# Step 2: Compute Confidence Interval
def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error = s / sqrt(n)

    # Use t-distribution for small samples (n < 30), otherwise normal Z-distribution
    if n >= 30:
        critical_value = stats.norm.ppf(1 - (1 - confidence) / 2)  # Z-score
    else:
        critical_value = stats.t.ppf(1 - (1 - confidence) / 2, df=n-1)  # t-score

    margin_of_error = critical_value * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    print(f"Sample Mean: {mean:.3f}")
    print(f"Confidence Level: {confidence*100}%")
    print(f"Margin of Error: {margin_of_error:.3f}")
    print(f"Confidence Interval: ({lower_bound:.3f}, {upper_bound:.3f})")

    return mean, lower_bound, upper_bound

mean, lower_bound, upper_bound = confidence_interval(sample_data)

# Step 3: Visualization
plt.figure(figsize=(8, 5))
plt.hist(sample_data, bins=15, color='lightblue', edgecolor='black', alpha=0.7, density=True)

# Plot mean and confidence interval
plt.axvline(mean, color='red', linestyle='--', label=f"Sample Mean: {mean:.2f}")
plt.axvline(lower_bound, color='green', linestyle='--', label=f"Lower Bound: {lower_bound:.2f}")
plt.axvline(upper_bound, color='green', linestyle='--', label=f"Upper Bound: {upper_bound:.2f}")

# Labels and legend
plt.xlabel("Data Values")
plt.ylabel("Density")
plt.title(f"Confidence Interval for Mean (95%)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
# 22. Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define normal distribution parameters
mu = 100  # Mean
sigma = 15  # Standard Deviation

# Generate x values (range for plotting)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Calculate PDF values
pdf_values = stats.norm.pdf(x, loc=mu, scale=sigma)

# Plot the PDF
plt.figure(figsize=(8, 5))
plt.plot(x, pdf_values, color='blue', linewidth=2, label=f"N({mu}, {sigma}²)")
plt.xlabel("X-values")
plt.ylabel("Probability Density")
plt.title("Probability Density Function (PDF) of Normal Distribution")
plt.axvline(mu, color='red', linestyle='--', label="Mean (μ)")
plt.axvline(mu - sigma, color='green', linestyle='--', label="μ - σ")
plt.axvline(mu + sigma, color='green', linestyle='--', label="μ + σ")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

# Show plot
plt.show()
```

```python
# 23. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define Poisson distribution parameter (λ = mean number of events)
lambda_value = 5

# Generate range of k values (discrete)
k_values = np.arange(0, 15)  # Poisson is discrete, so we use integer values

# Calculate CDF values
poisson_cdf = stats.poisson.cdf(k_values, mu=lambda_value)

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(k_values, poisson_cdf, where="mid", color='blue', label=f"Poisson(λ={lambda_value})", linewidth=2)
plt.scatter(k_values, poisson_cdf, color='red', zorder=3)
plt.xlabel("Number of Events (k)")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Distribution Function (CDF) of Poisson Distribution")
plt.axhline(1, color='gray', linestyle="--", alpha=0.7)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

# Show plot
plt.show()

# Interpretation: Probability of k ≤ 3 events occurring
k_value = 3
probability = stats.poisson.cdf(k_value, mu=lambda_value)
print(f"Probability of observing ≤ {k_value} events: {probability:.4f}")
```

```python
# 27. Implement Z-score normalization on a dataset and visualize its transformation

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a dataset with a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=1000)  # Mean=50, SD=15

# Compute mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data, ddof=1)  # Sample standard deviation

# Apply Z-score normalization
z_scores = (data - mean) / std_dev

# Plot original and normalized distributions
plt.figure(figsize=(10, 5))

# Original data distribution
plt.subplot(1, 2, 1)
sns.histplot(data, bins=30, kde=True, color="blue", alpha=0.7)
plt.axvline(mean, color='red', linestyle='--', label=f"Mean = {mean:.2f}")
plt.xlabel("Original Values")
plt.ylabel("Frequency")
plt.title("Original Distribution")
plt.legend()

# Normalized data distribution
plt.subplot(1, 2, 2)
sns.histplot(z_scores, bins=30, kde=True, color="green", alpha=0.7)
plt.axvline(0, color='red', linestyle='--', label="Mean = 0")
plt.xlabel("Z-score Values")
plt.ylabel("Frequency")
plt.title("Z-score Normalized Distribution")
plt.legend()

# Show plot
plt.tight_layout()
plt.show()

# Print summary statistics
print(f"Original Mean: {mean:.2f}, Original Std Dev: {std_dev:.2f}")
print(f"After Normalization: Mean = {np.mean(z_scores):.2f}, Std Dev = {np.std(z_scores, ddof=1):.2f}")
```
